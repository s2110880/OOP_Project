```python
from DataLoaders import read_csv, read_statsmodels, readCSV_fromWEB
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

Using designed read_csv class to load csv from local drive

```python
csv = read_csv(X = ["wool", "tension"], y = "breaks")
csv.data_loading(path="warpbreaks.csv", X = ["wool", "tension"], y = "breaks")
#Adding constant
csv.add_constant()
print(csv.X)
#Transposing
print(csv.x_transpose)

#printing y
print(csv.y)
```

Using designed csvWEB to read csv from web

```python
csvWEB = readCSV_fromWEB(X = ["wool", "tension"], y = "breaks")
url = "https://raw.githubusercontent.com/BI-DS/GRA-4152/refs/heads/master/warpbreaks.csv"
csvWEB.data_loading(url=url, X = ["wool", "tension"], y = "breaks")

#Adding constant
csvWEB.add_constant()
print(csvWEB.X)
#Transposing
print(csvWEB.x_transpose)

#printing y
print(csvWEB.y)
```
